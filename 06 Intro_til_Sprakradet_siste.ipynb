## 4. Korpusbygging

Korpuset er en definisjon av et korpus, og består av en samling URN-er.
Et korpus kan bygges rundt et tema, den dewey, en forfatter, en periode etc.

nb.book_urn() gir en samling av urn'er ut fra gitte spesifikasjoner. % er trunkering. 

Her lager vi et korpus med navnet urns for bøker av Anne Kath Vestly, 10 stk.

Korpuset kan vi så foreta ulike undersøkelser av:
    1. sttr
    2. ordfrekvens
    3. frekvens for spesifikke termer
    4. setningsuttrekk
    5. kollokasjoner 
    6. konkordanser
    
    
Til slutt viser vi forskjellige måter å bygge korpus på, og hvordan vi kan eksportere et korpus til excel,
redigere det, og hente det tilbake til notebook.

```python
urns = nb.book_urn(words=None,
    author='Vestly%Anne%',
    title=None,
    subtitle=None,
    ddk=None,
    subject=None,
    period=(1100, 2020),
    gender=None,
    lang=None,
    trans=None,
    limit=10,)
```

```python
urns
```

### 4.1. STTR

```python
## Standardized text token ratio. Måle lettlesthet. nb.sttr

res =[]

for x in urns: 
    res.append([nb.sttr(x[0])]+x)
    
nb.frame(res, 'sttr urn forfatter tittel år'.split())
```

### 4.2. Frekvens for ulike bøker

```python
# Her sammenlikner vi frekvenser for ulike bøker. I dette eksempelet Vestlys bøker.

# nb.get_corpus_text()

vestly = nb.get_corpus_text(urns)
vestly.head()
```

```python
frame_sort = lambda df, koll=0, antall=10: df.sort_values(by=df.columns[koll], ascending=False).head(antall)
```

```python
frame_sort(vestly, 1, 20)
```

```python
# Vi kan gjøre det samme, men legge på et varmekart for lettere å få oversikt

nb.heatmap(vestly.head(20))
```

### 4.3. Frekvens for spesifikke termer

```python
def heatmap_loc(dr, ordliste,sorter=0):
    df = dr.loc[ordliste]
    return nb.heatmap(df.sort_values(by=df.columns[sorter], ascending=False))
```

```python
# Her lager vi en funksjon heatmap_loc som vi bruker for å plukke ut enkelte ord fra korpuset som vi vil sammenlikne.


heatmap_loc(vestly, "Mormor Morten Marte Ole Aleksander Guro Lillebror Knerten".split())
```

```python
heatmap_loc(vestly, "boken boka sola solen".split())
```

### 4.4. Setningsuttrekk

```python
# nb.sentences() gir oss et utvalg setninger fra et korpus

nb.sentences(urns)
```

### 4.5. Kollokasjoner

```python
#nb.urn_coll() finner hvilke ord et ord er kollokert med. Kan bestemme selv hvor mange ord man vil ha før og etter.
nb.urn_coll('Mormor', urns=urns, after=2, before=2, limit=1000)
```

### 4.6. Konkordans

```python
#nb.urn_concordance er en måte å vise konkordanser på

nb.urn_concordance(urns=urns, word='Mormor', before=8, after=8)
```
